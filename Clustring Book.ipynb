```python
import pandas as pd 
import numpy as np 
```

```python
books=pd.read_csv('Books.csv')
users=pd.read_csv('Users.csv')
ratings=pd.read_csv('Ratings.csv')
```

```python
books.head(2)
```

```python
books=books[['ISBN','Book-Title','Book-Author','Year-Of-Publication','Publisher']]
```

```python
books.head(2)
```

```python
books.rename(columns={'Book-Title':'title','ISBN':'isbn','Book-Author':'author','Year-Of-Publication':'year','Publisher':'Publisher'},inplace=True)
```

```python
books.head()
```

```python
users.rename(columns={'User-ID':'user_id','Location':'location','Age':'age'},inplace=True)
```

```python
users.head(2)
```

```python
ratings.head()
```

```python
ratings.rename(columns={'User-ID':'user_id','Book-Rating':'rating'},inplace=True)
```

```python
ratings.head(1)
```

```python
books.shape
```

```python
users.shape
```

```python
ratings.shape
```

```python
x=ratings['user_id'].value_counts()>200
```

```python
y=x[x].index
```

```python
y
```

```python
ratings=ratings[ratings['user_id'].isin(y)]
```

```python
ratings
```

```python
ratings_with_books=ratings.merge(books,on='isbn')
```

```python
ratings_with_books.shape
```

```python
 number_rating=ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
number_rating.rename(columns={'rating':'no of rating'},inplace=True)
```

```python
number_rating
```

```python
final_rating=ratings_with_books.merge(number_rating,on='title')
```

```python
final_rating
```

```python
final_rating=final_rating[final_rating['no of rating']>50]
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(['user_id','title'],inplace=True)
```

```python
final_rating.shape
```

```python
book_pivot=final_rating.pivot_table(columns='user_id',index='title',values='rating')
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix 
book_sparse=csr_matrix(book_pivot)
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm='brute')
```

```python
book_sparse
```

```python
model.fit(book_sparse)
```

```python
distances,suggestions=model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1),n_neighbors=6)
```

```python
distances
```

```python
suggestions
```

```python
for i in range(len(suggestions)):
    print(book_pivot.index[suggestions[i]])
```

```python
book_pivot.index[237]
```

```python
def recomend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    
    
    distances, suggestions = model.kneighbors(
        book_pivot.iloc[book_id, :].values.reshape(1, -1),
        n_neighbors=6
    )
    

    for i in suggestions[0][1:]:
        print(book_pivot.index[i])
```

```python
recomend_book('Animal Farm')
```

```python
import pickle

# Save KNN model
with open("model.pkl", "wb") as f:
    pickle.dump(model, f)

# Save pivot table
with open("book_pivot.pkl", "wb") as f:
    pickle.dump(book_pivot, f)

# (Optional) Save books dataframe
with open("books.pkl", "wb") as f:
    pickle.dump(books, f)

# (Optional) Save ratings dataframe
with open("ratings.pkl", "wb") as f:
    pickle.dump(ratings, f)
```
